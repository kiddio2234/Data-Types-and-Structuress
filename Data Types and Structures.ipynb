17. Describe the time complexity of accessing elements in a dictionary.

--> Python dictionaries are implemented using hash tables.
Accessing a value by key (e.g. my_dict['name']) is very fast because it uses the key’s hash to find the value directly.

Many keys have the same hash (hash collisions).
Python has to scan multiple entries degrading performance.

However Python handles collisions efficiently so worst-case time is very rare.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

--> Dictionaries are called unordered because you don’t get items by position like in a list.

Instead, you get values using keys (like my_dict['name']) so the order doesn’t matter for finding data.

They are still not designed for position-based access (like lists).
You retrieve values by key, not by position.
So, the order doesn’t affect data retrieval, as lookup is always by key, not index.
